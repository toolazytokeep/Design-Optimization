## Problem 1

$$
\begin{aligned}
\min_{x_1,x_2} & \quad f({\bf x})=(x_1+1)^2+(x_2-2)^2\\
{\text{subject to }} & \quad g_1 = x_1-2\leq 0,{\quad} g_3 = -x_1\leq 0,\\
& \quad g_2 = x_2-1\leq 0, {\quad} g_4 = -x_2\leq 0.
\end{aligned}
$$

Find the optimum graphically. Determine directions of feasible descent at the corner points of the feasible domain. Show the gradient directions of $f$ and $g_i$s at these points. Verify graphical results analytically using the KKT conditions.

## Solution 1

We calculate the gradient descent for $f$ and $g_i$ at the conrner points:
$$
\begin{aligned}
\quad & \frac{\partial f}{\partial x}|_{x_1=0,x_2=0}=
\begin{bmatrix}
2(x_1 + 1) \\
2(x_2 -2) \\
\end{bmatrix}=
\begin{bmatrix}
2 \\
-4 \\
\end{bmatrix}\\
\quad & \frac{\partial f}{\partial x}|_{x_1=2,x_2=0}=
\begin{bmatrix}
2(x_1 + 1) \\
2(x_2 -2) \\
\end{bmatrix}=
\begin{bmatrix}
6 \\
-4 \\
\end{bmatrix}\\
\quad & \frac{\partial f}{\partial x}|_{x_1=0,x_2=1}=
\begin{bmatrix}
2(x_1 + 1) \\
2(x_2 -2) \\
\end{bmatrix}=
\begin{bmatrix}
2 \\
-2 \\
\end{bmatrix}\\
\quad & \frac{\partial f}{\partial x}|_{x_1=2,x_2=1}=
\begin{bmatrix}
2(x_1 + 1) \\
2(x_2 -2) \\
\end{bmatrix}=
\begin{bmatrix}
6 \\
-2 \\
\end{bmatrix}\\
\end{aligned}
$$

To find the gradient directions for $g_is$

$$
\begin{aligned}
\quad & \frac{\partial g_1}{\partial x}=
\begin{bmatrix}
1 \\
0 \\
\end{bmatrix}\\
\quad & \frac{\partial g_2}{\partial x}=
\begin{bmatrix}
0 \\
1 \\
\end{bmatrix}\\
\quad & \frac{\partial g_3}{\partial x}=
\begin{bmatrix}
-1 \\
0 \\
\end{bmatrix}\\
\quad & \frac{\partial g_4}{\partial x}=
\begin{bmatrix}
0 \\
-1 \\
\end{bmatrix}\\
\end{aligned}
$$

Finding the analytical solution using KKT conditions,

We use objective function and constraint to list the Lagrange expression with four parameter: 
$\mu_1, \mu_2, \mu_3, \mu_4$ 
with the conditions that : $\mu_1 \geq 0, \mu_2 \geq 0, \mu_3 \geq 0, \mu_4 \geq 0$

Hence the lagrangian expression can be defined as :

$$
\begin{aligned}
\quad & L(x_1, x_2, \lambda) = (x_1 + 1)^2 + (x_2 - 2)^2 + \mu_1(x_1 - 2)  + \mu_2(x_2 - 1) - \mu_3x_1 - \mu_4x_2
\end{aligned}
$$

Now, the differential of lagrangian expression is given as : 
$$
\begin{aligned}
\nabla_x \mathcal{L} =
\begin{bmatrix}
2(x_1 + 1) + \mu_1 - \mu_3 \\
2(x_2 - 2) + \mu_2 - \mu_4 \\
\end{bmatrix} = 
\begin{bmatrix}
0 \\
0 \\
\end{bmatrix}\\
\end{aligned}
$$

We guess that $\mu_2>0, \mu_3>0, \mu_1=\mu_4=0$, then 
$$
\begin{aligned}
\quad & x_1 = 0\\
\quad & x_2 = 1\\
\quad & \mu_3 = 2 > 0\\
\quad & \mu_2 = 2 > 0\\
\end{aligned}
$$

Since the objective function is convex function and feasible domain is convex set, the problem is convex problem. We can know the solution is 
$$
\begin{aligned}
\quad & x_1 = 0 \\
\quad & x_2 = 1 \\
\quad & \mu_1 = 0 \\ 
\quad & \mu_2 = 2 \\ 
\quad & \mu_3 = 2 \\
\quad & \mu_4 = 0 \\
\end{aligned}
$$

The minimum value for the objective function is 1.

Next, we guess that $\mu_1>0, \mu_4>0, \mu_2=\mu_3=0$, then 
$$
\begin{aligned}
\quad & x_1 = 2\\
\quad & x_2 = 0\\
\quad & \mu_1 = -6 < 0\\
\quad & \mu_4 = -4 < 0\\
\end{aligned}
$$

which $\mu_1$ and $\mu_4$ contradicts the assumption, so $x_1=2, x_2=0$ is not the minimum solution. From the figure attached below, we can see that this point should be the maximum solution for the problem.

Next, we guess that $\mu_1>0, \mu_2>0, \mu_3=\mu_4=0$, then 
$$
\begin{aligned}
\quad & x_1 = 2\\
\quad & x_2 = 1\\
\quad & \mu_1 = -6 < 0\\
\quad & \mu_2 = 2 > 0\\
\end{aligned}
$$

which $\mu_1$ contradicts the assumption, so $x_1=2, x_2=1$ is not the minimum solution. 

At last, we guess that $\mu_3>0, \mu_4>0, \mu_1=\mu_2=0$, then 
$$
\begin{aligned}
\quad & x_1 = 0\\
\quad & x_2 = 0\\
\quad & \mu_3 = 2 > 0\\
\quad & \mu_4 = -4 < 0\\
\end{aligned}
$$

which $\mu_4$ contradicts the assumption, so $x_1=0, x_2=0$ is not the minimum solution.

In summary, the minimum value for the objective function is 1 when $x_1=0, x_2=1$.

## Solution 3

$$
\begin{aligned}
\min_{x_1,x_2,x_3} & \quad  f=-x_1x_2-x_2x_3-x_1x_3\\
{\text{subject to }} & \quad h=x_1+x_2+x_3-3=0.
\end{aligned}
$$ 

Using reduced gradient method:

We set $x_1$ as the state varialbe, $x_2$ and $x_3$ as the decision variable. Then we can have 
$$
\begin{aligned}
\quad & \frac{{\partial}f}{{\partial}d} = 
\begin{bmatrix}
-x_{1} - x_{3} \\
-x_{1} - x_{2} \\
\end{bmatrix}\\
\quad & \frac{{\partial}f}{{\partial}s} = -x_2 - x_3\\
\quad & \frac{{\partial}h}{{\partial}d} = 
\begin{bmatrix}
1 \\
1 \\
\end{bmatrix}\\
\quad & \frac{{\partial}h}{{\partial}s} = 1\\
\end{aligned}
$$

It is easy to know that
$$
\begin{aligned}
\quad & \frac{{\rm d}s}{{\rm d}d} = -(\frac{{\partial}h}{{\partial}s})^{-1}(\frac{{\partial}h}{{\partial}d}) = 
\begin{bmatrix}
-1 \\
-1 \\
\end{bmatrix}\\
\end{aligned}
$$

$$
\begin{aligned}
\quad & \frac{{\rm d}f}{{\rm d}d} = \frac{{\partial}f}{{\partial}d} - \frac{{\partial}f}{{\partial}s}(\frac{{\partial}h}{{\partial}s})^{-1}(\frac{{\partial}h}{{\partial}d}) 
= 
\begin{bmatrix}
-x_{1} - x_{3} \\
-x_{1} - x_{2} \\
\end{bmatrix}
- 
(-x_{2} - x_{3})
\begin{bmatrix}
1 \\
1 \\
\end{bmatrix} =
\begin{bmatrix}
-x_1 + x_2 \\
-x_1 + x_3 \\
\end{bmatrix} =
\begin{bmatrix}
0 \\
0 \\
\end{bmatrix}\\
\end{aligned}
$$

Hence, we can have $x_1=x_2=x_3$, plug the solution into the constraint $h=x_1+x_2+x_3-3=0$, then $x_1=x_2=x_3=1$. Next we need to consider the second-order condition, which is
$$
\begin{aligned}
\quad & \frac{{\partial}^2 h}{{\partial}d^2} = 
\begin{bmatrix}
0 & 0 \\
0 & 0 \\
\end{bmatrix}\\
\quad & \frac{{\partial}^2 h}{{\partial}s^2} = 0\\
\quad & \frac{{\partial}^2 h}{{\partial}s{\partial}d} = 
\begin{bmatrix}
0 & 0 \\
\end{bmatrix}\\
\quad & \frac{{\partial}^2 h}{{\partial}d{\partial}s} = 
\begin{bmatrix}
0 \\
0 \\
\end{bmatrix} \\
\end{aligned}
$$

$$
\begin{aligned}
\quad & \frac{{\rm d}^2 h}{{\rm d}d^2} = 
\begin{bmatrix}
I&(\frac{{\rm d}s}{{\rm d}d})^T\\
\end{bmatrix}
\begin{bmatrix}
\frac{{\partial}^2 h}{{\partial}d^2} & \frac{{\partial}^2 h}{{\partial}d{\partial}s}\\
\frac{{\partial}^2 h}{{\partial}s{\partial}d} & \frac{{\partial}^2 h}{{\partial}s^2}\\
\end{bmatrix}
\begin{bmatrix}
I \\
\frac{{\rm d}s}{{\rm d}d}\\
\end{bmatrix}
+
\frac{{\partial}h}{{\partial}s}\frac{{\rm d}^2 s}{{\rm d}d^2} = 0\\
\end{aligned}
$$

We can derive that $\frac{{\rm d}^2 s}{{\rm d}d^2} = 0$, then
$$
\begin{aligned}
\quad & \frac{{\partial}^2 f}{{\partial}d^2} = 
\begin{bmatrix}
0 & -1 \\
-1 & 0 \\
\end{bmatrix}\\
\quad & \frac{{\partial}^2 f}{{\partial}s^2} = 0\\
\quad & \frac{{\partial}^2 f}{{\partial}s{\partial}d} = 
\begin{bmatrix}
-1 & -1 \\
\end{bmatrix}\\
\quad & \frac{{\partial}^2 f}{{\partial}d{\partial}s} = 
\begin{bmatrix}
-1 \\
-1 \\
\end{bmatrix}\\
\end{aligned}
$$

$$
\begin{aligned}
\quad & \frac{{\rm d}^2 f}{{\rm d}d^2} = 
\begin{bmatrix}
I&(\frac{{\rm d}s}{{\rm d}d})^T\\
\end{bmatrix}
\begin{bmatrix}
\frac{{\partial}^2 f}{{\partial}d^2} & \frac{{\partial}^2 f}{{\partial}d{\partial}s}\\
\frac{{\partial}^2 f}{{\partial}s{\partial}d} & \frac{{\partial}^2 f}{{\partial}s^2}\\
\end{bmatrix}
\begin{bmatrix}
I \\
\frac{{\rm d}s}{{\rm d}d}\\
\end{bmatrix}
+
\frac{{\partial}f}{{\partial}s}\frac{{\rm d}^2 s}{{\rm d}d^2} =
\begin{bmatrix}
1&-1&-1 \\
\end{bmatrix}
\begin{bmatrix}
0&-1&-1 \\
-1&0&-1 \\
-1&-1&0 \\
\end{bmatrix}
\begin{bmatrix}
1 \\
-1 \\
-1 \\
\end{bmatrix} = 2 > 0\\
\end{aligned}
$$

Therefore, we know that when $x_1=x_2=x_3=1$, the objective function $f=x_1x_2+x_2x_3+x_1x_3$ has maximum value, which is 3.


## Solution 4

By implementing the generalized reduced gradient algorithm, we need to find the solution for the given function :

$$
\begin{aligned}
\min_{x_1,x_2,x_3} & \quad  f=x_{1}^{2}+x_{2}^{2}+x_{3}^{2}\\
{\text{subject to }} & \quad h_1 = x_{1}^{2}/4+x_{2}^{2}/5+x_{3}^{2}/25-1=0\\
& \quad h_2 = x_1+x_2-x_3= 0,
\end{aligned}
$$ 

```python
import numpy as np

def obj(x1, x2, x3):
    return x1 ** 2 + x2 ** 2 + x3 ** 2

def df_dx(x1, x2, x3):
    df_ds = np.vstack((2 * x2, 2 * x3))
    df_dd = np.array([2 * x1], dtype=float)
    return df_ds, df_dd

def constraint(x1, x2, x3):
    h1 = x1 ** 2 / 4 + x2 ** 2 / 5 + x3 ** 2 / 25 - 1
    h2 = x1 + x2 - x3
    return h1, h2

def dh_dx(x1, x2, x3):
    dh_ds = np.vstack((np.hstack(((2/5) * x2, (2/25) * x3)), np.array([[1, -1]], dtype=float)))
    # dh_ds = np.array([[(2 / 5) * x2, (2 / 25) * x3], [1, -1]], dtype=float)
    dh_ds_inv = np.linalg.inv(dh_ds)
    dh_dd = np.vstack(((1 / 2) * x1, 1.))
    return dh_ds_inv, dh_dd, dh_ds

def dfdd(x1, x2, x3):
    df_ds, df_dd = df_dx(x1, x2, x3)
    dh_ds_inv, dh_dd, _ = dh_dx(x1, x2, x3)
    return df_dd - np.matmul(df_ds.T, np.matmul(dh_ds_inv, dh_dd))

def solveh(x1, x2, x3):
    error = 1e-3
    # check if the constraint is satisfied
    h1, h2 = constraint(x1, x2, x3)
    h = np.vstack((h1, h2))
    h_norm = np.linalg.norm(h)
    # take pseudo inverse
    while h_norm >= error:
        dh_inv, _, dh = dh_dx(x1, x2, x3)
        Lambda = 1
        ds = np.matmul(dh_inv, h)
        # ds = np.matmul(np.matmul(np.linalg.inv(np.matmul(dh.T, dh) + Lambda * np.array([[1, 0], [0, 1]])), dh.T), h).flatten()
        x2 = x2 - ds[0]
        x3 = x3 - ds[1]
        h1, h2 = constraint(x1, x2, x3)
        h = np.vstack((h1, h2))
        h_norm = np.linalg.norm(h)
    return x1, x2, x3, h_norm

def line_search(x1, x2, x3):
    a = 1.  # initialize step size
    df = dfdd(x1, x2, x3)
    phi = lambda a, x1, x2, x3, df: obj(x1, x2, x3) - a * 0.3 * np.matmul(df, df.T)  # define phi as a search criterion

    def f_a(x1, x2, x3, a):
        df = dfdd(x1, x2, x3)
        dh_ds_inv, dh_dd, _ = dh_dx(x1, x2, x3)

        x1 = x1 - a * df.flatten()
        ds = np.matmul(np.matmul(dh_ds_inv, dh_dd), df.T).flatten()
        x2 = x2 + a * ds[0]
        x3 = x3 + a * ds[1]
        return obj(x1, x2, x3)

    while phi(a, x1, x2, x3, df) < f_a(x1, x2, x3, a):
        a = 0.5 * a
        df = dfdd(x1, x2, x3)
    return a

eps = 1e-3  # termination criterion
x1 = np.array([1.0], dtype=float)
x2 = np.array([2.0], dtype=float)
x3 = np.array([3.0], dtype=float)
iter = 0  # counter

x1, x2, x3, _ = solveh(x1, x2, x3)

df_dd_norm = np.linalg.norm(dfdd(x1, x2, x3))

while df_dd_norm >= eps:  # keep searching while gradient norm is larger than eps
    a = line_search(x1, x2, x3)
    x1 = x1 - a * dfdd(x1, x2, x3).flatten()
    dh_ds_inv, dh_dd, dh = dh_dx(x1, x2, x3)
    ds = a * np.matmul(np.matmul(dh_ds_inv, dh_dd), dfdd(x1, x2, x3).T).flatten()
    x2 = x2 + ds[0]
    x3 = x3 + ds[1]
    x1, x2, x3, _ = solveh(x1, x2, x3)
    df_dd_norm = np.linalg.norm(dfdd(x1, x2, x3))

print('Solution is: X= ', x1,x2,x3)
```

## Problem 5

Garbage truck routing problem,

Let there be N sites to be visited and consider them as nodes of a graph. The cost of moving from node
i to j is $c_{ij}$ if there is an edge between the nodes, or  if there is none. 

Site 0 is the truck station where the truck starts and returns.

Formulate the problem to minimize the total cost while the truck visits all sites and
returns to the station.


## Solution

The problem can be formulated as:
$${
\text{min} \sum_{i=0}^{N}\sum_{j=0}^{N} c_{ij}x_{ij}
}$$

To constrain the route to go exactly to other site different from the one we just left:
$${
\sum_{i=0}^{N} x_{ij} = 1, \quad \text{for}\quad j = 0,1,2,...N \quad \text{and} \quad i\neq j
}$$
To constrain the route to go exactly to other site different from the one we just arrives:
$${
\sum_{k=0}^{N} x_{jk} = 1, \quad \text{for}\quad j = 0,1,2,...N \quad \text{and} \quad k\neq j
}$$

Then the complete problem formulation can be given as:
$${
\text{min} \sum_{i=0}^{N}\sum_{j=0}^{N} c_{ij}x_{ij}
}$$
